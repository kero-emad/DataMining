```python
from sklearn.preprocessing import MultiLabelBinarizer
import pandas as pd
```

```python
df=pd.read_csv('survey_results_public_after_cleaning.csv')
```

```python
columns_to_encode = [
    'LanguageHaveWorkedWith',
    'DatabaseHaveWorkedWith',
    'PlatformHaveWorkedWith',
    'WebframeHaveWorkedWith',
    'MiscTechHaveWorkedWith',
    'NEWCollabToolsHaveWorkedWith'
]
```

```python
encoded_dataframes = []
for column in columns_to_encode:
    df[column] = df[column].apply(lambda x: list(set(x.split(';'))))
    mlb = MultiLabelBinarizer()
    one_hot_encoded = mlb.fit_transform(df[column])
    encoded_df = pd.DataFrame(one_hot_encoded, columns=[f"{column}_{val}" for val in mlb.classes_], index=df.index)
    encoded_dataframes.append(encoded_df)

final_encoded_df = pd.concat(encoded_dataframes, axis=1)

final_encoded_df['DevType'] = df['DevType']


```
